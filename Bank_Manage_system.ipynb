Account Class (Base Class)

Attributes: account_number, name, balance, account_type

Methods: deposit(), withdraw(), check_balance()

SavingAccount & CurrentAccount (Derived Classes)

SavingAccount: Minimum balance, interest calculation

CurrentAccount: Overdraft limit

Bank Class (Manager Class)

Manage multiple accounts

Methods: add_account(), remove_account(), find_account(), transfer_money()

Transaction Logging

Keep track of all deposits, withdrawals, and transfers

Optional: Store logs in a .txt or .csv file for persistence

Admin & Customer Menu

Admin: Add/Remove accounts, View all accounts

Customer: Deposit, Withdraw, Transfer, View Balance

```python
id = 0 
database = {'Normal Account':{},
            'Current':{},
            'Saving Account':{}}
managerdatabase = {}

class Customers : 
    def __init__(self,name,balance=0,account_type='Normal Account'):
        global id 
        self.accountNumber = id
        id+=1
        self.name = name
        self.account_type = account_type
        self.balance = balance
        # Storing Data In Data Base
        database[account_type][self.accountNumber] = {'Name':self.name,'Balance':self.balance}
        self.Greet_User()
    def Greet_User(self):
        print(f'Welcome User {self.name} You Have Started Your Account With Balance {self.balance}')
    def Check_Balance(self):
        print(f'Your Balance Is {self.balance}')
    def add_amount(self,amount):
        print(f'You Have Succesfully Added {amount} in your Bank Account')
        database[self.account_type][self.accountNumber]['Balance'] += amount
        self.balance += amount
        print(f'Total Balance {self.balance}')
    def withdraw_amount(self,amount):
        if amount > self.balance : 
            print(f'The Amount You Entered Is Greater Then Your Accounts Balance.')
        else : 
            self.balance -= amount
            # Updating In The Data Base
            database[self.account_type][self.accountNumber]['Balance'] -= amount
            print(f'The {amount} Has Been Deducted From Your Account and your current balance is {self.balance} .')
    def transfer_money(self,amount,reciver,acount_type,reciver_id):
        if amount>self.balance:
            print(' The Amount You Want To Send Is Greater Then Your Balance')
            return
        if reciver_id in database[acount_type].keys() :
            # Updating In Self Of Sender 
            self.balance -= amount
            # Updating In Database Of Sender 
            database[self.account_type][self.accountNumber]['Balance'] -= amount
            # Updating In Reciver Self
            reciver.balance += amount
            # Updating in Reciver Data Base
            database[acount_type][reciver_id]['Balance'] += amount
        else : 
            print('Invalid Crenditals') 
class Manager :
    global database 
    def __init__(self,name,):
        self.name = name 
        # Updating In Data Base 
        managerdatabase[self.name] = {'Name':self.name}
        self.greet_manager()
    def greet_manager(self) : 
        print(f'Hello {self.name} ! Manager You Are Appointed As A Manager In SMIT BANK.PK ')
    def view_all_coustomers(self):
        for i in database:
            print(f'Account Type {i} ')
            for j,k in database[i].items() : 
                print(f"Id, {j} Name {k['Name']} \nBalance {k['Balance']}")
                print('----------------------------------------------')
    def delete_customers(self,account_number_to_delete,account_type) :
        if account_number_to_delete in  database[account_type] : 
            i = database[account_type][account_number_to_delete]
            del database[account_type][account_number_to_delete]
            print(f"The User With Name {i['Name']} and ID {account_number_to_delete} Has Been Sucesfully Deleted.")
        else : 
            print('Invalid Crenditals')
        
```
